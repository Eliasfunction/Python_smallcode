```python
import requests
import json
import pandas
url = "https://ecshweb.pchome.com.tw/search/v3.3/all/category/DRAH/results?q=ssd&page=1&sort=sale/dc"
#this url is pchome24H ,Keyword is "ssd"
#change url's page you can get other Commodity Information, just a "for"
res = requests.get(url)
res = json.loads(res.text)
webdatas = res['prods']
rr = pandas.DataFrame(webdatas)
#rr.to_excel('data_sample.xlsx')
rr
```

```python
for product in webdatas:
    print(product['name'])
    print('$',product['price'])
    print('https://b.ecimg.tw/'+product['picB'])
    print('https://24h.pchome.com.tw/prod/'+product['Id'])
    print('\n')
```

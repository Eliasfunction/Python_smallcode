```python
from selenium import webdriver
driver = webdriver.chrome()
driver.get('https://shopee.tw/')
```

```python
mycookies = ';'.join(['{}={}'.format(item.get('name'),item.get('value')) for item in driver.get_cookies()])
#mycookies
```

```python
import requests
import json

searchapi = 'https://shopee.tw/api/v4/search/search_items?'
payload = '?by=relevancy&keyword=%23%E7%8E%A9%E5%85%B7&limit=50&newest=0&order=desc&page_type=search&version=2'
api = searchapi + payload

headers = {
    'cookie': mycookies,
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
}

res = requests.get(payload,headers = headers)
#res.json()
```

```python
import pandas
pp = res.json()
pp = pp['items']
#pp
for items in pp:
    pc = items['item_basic']
    print('Name  : '+pc['name'])
    print('Image : https://cf.shopee.tw/file/'+pc['image'])
    print('Prod  : https://shopee.tw/product/'+str(pc['shopid'])+'/'+str(pc['itemid']))
    print(pc['currency']+'   : ', end = '')
    print(round(pc['price']/100000))
    print('Place : ', end = '')
    try:
        Place = pc['shop_location']
        print(Place)   
    except:
        print("  ")
  
    print(' ')
```

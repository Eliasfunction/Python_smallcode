```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver')
driver.get('https://shopee.tw/flash_sale?categoryId=0&promotionId=2009322110')
```

```python
ary = []
csrftoken =''
for stinfo in driver.get_cookies():
    if stinfo.get('name') == 'csrftoken':
        csrftoken = stinfo.get('value')
    ary.append('{}={}'.format(stinfo.get('name'), stinfo.get('value')))
cookie = ';'.join(ary)
```

```python
csrftoken
```

```python
import requests
import json
headers = {
    'x-csrftoken':csrftoken,
    'referer':'https://shopee.tw/flash_sale?promotionId=2009322110',
    'content-type':'application/json',
    'cookie':cookie
}
payload = json.loads('''{"promotionid":2009322110,"categoryid":0,"itemids":[2911689736,4748071521,4612228116,4778425569,2611989044,3546418194,6062919437,7779189659,2433248518,9417167872,5202591136,8305561859],"sort_soldout":true,"limit":12,"need_personalize":true,"with_dp_items":true}''')
res = requests.post('https://shopee.tw/api/v2/flash_sale/flash_sale_batch_get_items', json = payload, headers= headers)
#res.json()
```
